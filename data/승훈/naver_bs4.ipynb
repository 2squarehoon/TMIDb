```python
rating = pd.read_csv('data_naver.csv', encoding='utf-8')

for i in range(915, 916): # 37, 43 월E, 115, 530 건너뜀
    if rating['year'][i] != 0:
        title = rating['title'][i] # title string 뽑아오기
        url = 'https://movie.naver.com/movie/search/result.naver?query='
        q = title.replace(' ', '+').replace(':', '%3A').replace(',', '%2C')
        page = urlopen(url+q)
        soup = BeautifulSoup(page, 'html.parser')
        search_cnt = len(soup.find_all('em', 'num')) # 검색해서 나온 영화의 수
        for j in range(search_cnt): # 검색해서 나온 영화수만큼 반복문 돌려서 naver_rate가 같은 영화찾기
            rate = float(soup.find_all('em', 'num')[j].string) # 네이버 영화 페이지에서 점수값 뽑아오기, type이 달라서 실제값이 같아도 적용이 되지를 않는다.
            if rating['naver_rating'][i] and rate == rating['naver_rating'][i]: # 뽑아온 점수와 db안의 점수가 같으면
                rating['naver_count'][i] = int(soup.find_all('em', 'cuser_cnt')[j].string[4:-2]) # 평점 남긴 사람수 기록

            
rating = rating.set_index(keys=['titleId'], inplace=False)
rating.to_csv('data_naver.csv')
```
